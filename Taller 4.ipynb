### Define schema

```python
table = instance.table(TABLE_ID)

if not table.exists():
    table.create()
else:
    print("ERROR: Table {} already exists".format(TABLE_ID))
```

```python
GENERATION_CF_ID= 'generation'
generation_cf = table.column_family(GENERATION_CF_ID)
generation_cf.create()
```

```python
SENSOR_CF_ID= 'weather_sensor'
sensor_cf = table.column_family(SENSOR_CF_ID)
sensor_cf.create()
```

### Inserting data (Point 1)

```python
def insert_data(df, column_family, is_gen= True):
    rows= []
    for r in df.itertuples(index= False):
        #print(f'{r[1]};{r[2]};{r[0]}')
        row_key= f'{r[1]};{r[2]};{r[0]}'.encode()
        row= table.direct_row(row_key)
        if is_gen:
            #row.set_cell(column_family, 'DC_POWER', struct.pack('f', r[3]))
            #row.set_cell(column_family, 'AC_POWER', struct.pack('f', r[4]))
            row.set_cell(column_family, 'DAILY_YIELD'.encode(), str(r[5]))
            #row.set_cell(column_family, 'TOTAL_YIELD', struct.pack('f', r[6]))
        else:
            row.set_cell(column_family, 'AMBIENT_TEMPERATURE'.encode(), str(r[3]))
            #ow.set_cell(column_family, 'MODULE_TEMPERATURE', struct.pack('f', r[4]))
            #row.set_cell(column_family, 'IRRADIATION', struct.pack('f', r[5]))
        #row.commit()
        rows.append(row)
    return rows
```

```python
table.mutate_rows(insert_data(p1_gen, GENERATION_CF_ID))
table.mutate_rows(insert_data(p1_sen, SENSOR_CF_ID, is_gen= False))
table.mutate_rows(insert_data(p2_gen, GENERATION_CF_ID))
table.mutate_rows(insert_data(p2_sen, SENSOR_CF_ID, is_gen= False))
```

```python
def print_row(row):
    print("Reading data for {}:".format(row.row_key.decode('utf-8')))
    for cf, cols in sorted(row.cells.items()):
        print("Column Family {}".format(cf))
        for col, cells in sorted(cols.items()):
            for cell in cells:
                labels = " [{}]".format(",".join(cell.labels)) \
                    if len(cell.labels) else ""
                print(
                    "\t{}: {} @{}{}".format(col.decode('utf-8'),
                                            cell.value.decode('utf-8'),
                                            cell.timestamp, labels))
    print("")
```

### Point 2

```python
rows= table.read_rows(filter_= row_filters.RowFilterChain(
            filters=[
                row_filters.RowKeyRegexFilter("4136001;\w+;2020-06-01 (0[6-9]|11|12):\d{2}:\d{2}".encode("utf-8")),
                row_filters.FamilyNameRegexFilter(SENSOR_CF_ID.encode()),                
            ]
        )
    )
for row in rows:
    print_row(row)
```
